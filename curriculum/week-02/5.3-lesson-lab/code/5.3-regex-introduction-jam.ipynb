```python
import pandas as pd
titanic_df = pd.read_csv("../assets/data/titanic.csv")
titanic_df.head(3)

pattern =  re.compile(r"\w+(?=,|\.)")
for q in titanic_df.Name: print q, pattern.findall(q)
```

```python
string = "tasdfasdfaj324-335-3412thereisl3.4etosaytodayaboutthesportingeventsconferenceroom8and9.2334.?#$}$"
pattern = re.compile(r'\d{0,3}[- \s]?\d{3,7}[- \s]?\d{4}')
pattern.findall(string)
```

```python
pattern = re.compile(r'-{0,1}\d+\.\d+')
pattern.findall(string)
```

```python
string2 = "The @quick %brown *fox"
pattern = re.compile(r'[\b\w+\s]')
pattern.findall(string2)
```

```python
pattern = re.compile(r'[\[0-9]{3}', re.I)
pattern.findall(string)
```
