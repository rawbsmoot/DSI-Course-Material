```python
import pandas as pd
import numpy as np
mnames = ['movie_id', 'title', 'genres']

#https://github.com/wesm/pydata-book/tree/master/ch02/movielens

movies = pd.read_table('https://raw.githubusercontent.com/wesm/pydata-book/master/ch02/movielens/movies.dat', sep='::', header=None,
                        names=mnames, engine='python')
movies.columns
```

```python
#Bretts solution
genre_List = gL = []
for i in movies.genres:
   Combined = i.split('|')
   for j in Combined:
       gL.append(j)
cleanedGenres = set(gL)
Combined

# gL
```
